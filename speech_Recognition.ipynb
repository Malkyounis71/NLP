```python
from gtts import gTTS
import langdetect
import IPython.display as ipd

# Function to generate audio from text
def text_to_speech(text):
    # Detect the language of the input text
    lang = langdetect.detect(text)

    # Create a gTTS object with the detected language
    tts = gTTS(text=text, lang=lang)

    # Save the speech to an audio file
    tts.save("output_audio.mp3")

    # Play the audio
    return ipd.Audio("output_audio.mp3")

# Input text (in any language)
input_text = "Je voudrais savoir quels sont les monuments historiques célèbres en France."

# Generate audio from the input text
audio = text_to_speech(input_text)

# Play the generated audio
audio
```
